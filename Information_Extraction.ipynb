# Pre Processing
Write a function that takes each document and performs:
1) sentence segmentation 2) tokenization 3) part-of-speech tagging

Please keep in mind that the expected output is a list within a list as shown below.


```python
def ie_preprocess(document):

    if isinstance(document, list):
        pos_sentences = nltk.pos_tag_sents(nltk.word_tokenize(sentence)
                                           for doc in document
                                           for sentence in nltk.sent_tokenize(doc))
    else:
        pos_sentences = nltk.pos_tag_sents(nltk.word_tokenize(sentence)
                                           for sentence in nltk.sent_tokenize(document))

    return pos_sentences
```

# Named Entity Extraction
Write a function that will take the list of tokens with POS tags for each sentence and returns the named entities (NE). 

Hint: Use binary=True while calling NE chunk function

```python
def named_entity_finding(pos_sent):

    named_entities = list()
    entities = nltk.ne_chunk(pos_sent, True)
    for subtree in entities.subtrees(filter=lambda t: t.label() == 'NE'):
        ne = ''
        for leaf in subtree.leaves():
            ne = ne + leaf[0] + ' '
        named_entities.append(ne.strip())

    return named_entities

pos_sents=ie_preprocess(list_of_doc)
named_entity_finding(pos_sents[0])
```

```python
def NE_flat_list_fn(pos_sents):
    NE=[]
    for pos_sent in pos_sents:
        NE.append(named_entity_finding(pos_sent))
    NE_flat_list = [ne for ne_list in NE for ne in ne_list]
    return NE_flat_list
```

### Generic Functions

Generic function to extract strings matching regular expression

```python
def extract_re(pattern, doc):
    pattern_compile = re.compile(pattern)
    words = pattern_compile.findall(doc)
    return words
```

Generic function to convert gazetteer (list of words) to regular expression pattern

```python
def get_re_patters_from_list(gazetteer):
    re_pattern = ''
    for position in gazetteer:
        re_pattern += position + '|'
    return "("+re_pattern[0:-1]+")"
```

### Name of the player

Extract name of the player by matching the first part of the string until ',' or '('

Note: Extracting Name of the player using named_entity is not giving the full name. This is bacause some of the non english names have words starting with lower case letters, for example 'dos', 'de' and special characters which are not captured by named entity chunker. For all the english names, ne_chunk is able to identify the full name. For this reason, I have used regular expression instead of ne_chunk.

```python
def name_of_the_player(doc):
    return extract_re(r"^.[^,(]*", doc)[0].strip()
```

### Country of Origin
Extracted the country of origin by extracting the word between 'the' and 'national team'

```python
def country_of_origin(doc):
    return extract_re(r"the (\w+)(?:\snational team)", doc)[0].strip()
```

### Date of Birth
Extracted Date of birth by extrating the text after 'born' and before the closing bracket.

```python
def date_of_birth(doc):
    return extract_re(r"(?:\(born\s)(.*?)(?:\))", doc)[0].strip()
```

# Convert to JSONLD

Write a function using the outputs from the previous functions to generate JSON-LD output as follows.

Reference: https://json-ld.org/primer/latest/

{ "@id": "http://my-soccer-ontology.com/footballer/name_of_the_player",

    "name": "",
    "born": "",
    "country": "",
    "position": [
        { "@id": "http://my-soccer-ontology.com/position",
            "type": ""
        }
     ]   
     "team": [
        { "@id": "http://my-soccer-ontology.com/team",
            "name": ""
        }   
     ]
}


```python
import json

def generate_jsonld(doc):
    player_json = dict()
    player_json["@id"] = "http://my-soccer-ontology.com/footballer/name_of_the_player"
    player_json["name"] = name_of_the_player(doc)
    player_json["born"] = date_of_birth(doc)
    player_json["country"] = country_of_origin(doc)
    player_json["position"] = list()
    player_json["team"] = list()
    positions = position_of_the_player(doc)
    for position in positions:
        player_json["position"].append({"@id": "http://my-soccer-ontology.com/position", "type": position})
    teams = team_of_the_player(doc)
    for team in teams:
        player_json["team"].append({"@id": "http://my-soccer-ontology.com/team", "name": team})
    return player_json
```

### FIFA awards won

FIFA awards won has been extracted for each player by extracting the sentence with word 'won ' or 'win ' in the text. Then a predefined list of awards is used to match against the extracted sentence to get the awards won.

Some of the players doesn't have any awards in the given text or haven't won any awards. For such players the list will be empty. 

```python
def fifa_awards_of_the_player(doc):
    awards_gazetteer = [
        "Ballon d'Or",
        "Ballons d'Or",
        "FIFA World Player of the Year",
        "FIFA Puskás Award",
        "FIFA Fair Play Award",
        "FIFA/FIFPro Best XI",
        "FIFA Best XI",
        "FIFPro Best XI",
        "FIFA Presidential Award"]

    awards_gazetteer_re = get_re_patters_from_list(awards_gazetteer)

    awards = list()
    for sent in extract_re(r"(?:won\s|win\s)(.*?)\.",doc):
        awards += extract_re(awards_gazetteer_re, sent)

    awards = list(set(awards))

    # To standardise the awards format convert Ballon d'Or or Ballons d'Or to FIFA Ballon d'Or
    fifa_ballon = False
    if "Ballon d'Or" in awards:
        awards.remove("Ballon d'Or")
        fifa_ballon = True
    if "Ballons d'Or" in awards:
        awards.remove("Ballons d'Or")
        fifa_ballon = True
    if fifa_ballon:
        awards.append("FIFA Ballon d'Or")

    return awards
```
